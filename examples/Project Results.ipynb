### Import Library

```python
# Import from parent directory
import sys
sys.path.insert(0,'..')

# This is what you would need to import
from osmpoi import *
```

### Getting Exeter Locations

```python
# Exeter City Centre - 250m radius
exeter_city = retrieve.pois_bbox(50.723274,-3.530833,50.727772,-3.523729)

# Exeter University Campus - 250m radius
exeter_uni = retrieve.pois_bbox(50.734646,-3.538690,50.739127,-3.531609)
```

### Getting Cardiff Locations

```python
# Cardiff City Centre - 250m radius
cardiff_city = retrieve.pois_bbox(51.477669,-3.180388,51.482159,-3.173177)

# Cardiff University Campus - 250m radius
cardiff_uni = retrieve.pois_bbox(51.485633,-3.183378,51.490119,-3.176174)
```

### Printing them as dictionaries with Natural Numbers

```python
# Confirming it works
print("Exeter City")
print(exeter_city)
print()  # Gaps just to make it look nicer
print("Exeter University")
print(exeter_uni)
print()
print("Cardiff City")
print(cardiff_city)
print()
print("Cardiff University")
print(cardiff_uni)
```

### As Percentages

```python
# Cardiff City Centre - 250m radius
cardiff_city_p = retrieve.pois_percent(51.477669,-3.180388,51.482159,-3.173177)

# Cardiff University Campus - 250m radius
cardiff_uni_p = retrieve.pois_percent(51.485633,-3.183378,51.490119,-3.176174)
```

```python
# Exeter City Centre - 250m radius
exeter_city_p = retrieve.pois_percent(50.723274,-3.530833,50.727772,-3.523729)

# Exeter University Campus - 250m radius
exeter_uni_p = retrieve.pois_percent(50.734646,-3.538690,50.739127,-3.531609)
```

### Printing them as dictionaries with Percentages

```python
# Confirming it works
print("Exeter City")
print(exeter_city_p)
print()  # Gaps just to make it look nicer
print("Exeter University")
print(exeter_uni_p)
print()
print("Cardiff City")
print(cardiff_city_p)
print()
print("Cardiff University")
print(cardiff_uni_p)
```

## Plotting

## Comparisons - Percentages

There is a third argument in the comparison called 'percent' this is set to 'True' by default

```python
plotter.compare(exeter_city, cardiff_city)
```

```python
plotter.compare(exeter_uni, cardiff_uni)
```

```python
plotter.compare(exeter_uni, cardiff_uni, percent=False)
```
